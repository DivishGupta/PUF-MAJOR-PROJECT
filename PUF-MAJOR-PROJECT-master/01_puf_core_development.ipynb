## IMPORT LIBRARIES

```python
# ==========================================================
# IMPORTS
# ==========================================================

import sys
sys.path.append("../")

import numpy as np
from src.puf_models import ArbiterPUF, XORPUF, generate_crps
```

## STEP 1 — Arbiter PUF (single chain)

```python
# ==========================================================
# Single Arbiter PUF — noise-free
# ==========================================================

arbiter = ArbiterPUF(n_stages=8, noise=0.0, seed=42)

# Generate one challenge and query response
challenge = arbiter.generate_challenge()
response  = arbiter.get_response(challenge)

print("Arbiter PUF:", arbiter)
print("Challenge :", challenge)
print("Response  :", response)
```

## STEP 2 — XOR Arbiter PUF

```python
# ==========================================================
# XOR Arbiter PUF — k=2, noise-free
# ==========================================================

xor_puf = XORPUF(n_stages=8, k=2, noise=0.0, seed=42)

challenge = xor_puf.generate_challenge()
response  = xor_puf.get_response(challenge)

print("XOR PUF  :", xor_puf)
print("Challenge:", challenge)
print("Response :", response)
```

## STEP 3 — XOR Arbiter PUF with Noise

```python
# ==========================================================
# XOR Arbiter PUF — k=2, Gaussian noise sigma=0.1
# ==========================================================

noisy_puf = XORPUF(n_stages=8, k=2, noise=0.1, seed=42)

challenge = noisy_puf.generate_challenge()
response  = noisy_puf.get_response(challenge)

print("Noisy XOR PUF:", noisy_puf)
print("Challenge    :", challenge)
print("Response     :", response)
```

## STEP 4 — CRP Dataset Generation (Sanity Check)

```python
# ==========================================================
# Generate 5 CRPs for a 2-XOR, 8-stage PUF
# ==========================================================

puf = XORPUF(n_stages=8, k=2, noise=0.0, seed=42)
X, y = generate_crps(puf, num_samples=5, seed=0)

print("Challenges:\n", X)
print("\nResponses:\n", y)
```
